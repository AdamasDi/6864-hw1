```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
test_50 = pd.read_csv('./test_eval_50.csv')
test_250 = pd.read_csv('./test_eval_250.csv')
test_500 = pd.read_csv('./test_eval_500.csv')
test_750 = pd.read_csv('./test_eval_750.csv')
test_1000 = pd.read_csv('./test_eval_1000.csv')
```

```python
test_50.head()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax[0][0].plot(test_50['n_train'], test_50['word'])
ax[0][0].plot(test_50['n_train'], test_50['lsa'])
ax[0][0].plot(test_50['n_train'], test_50['combo'])
ax[0][0].set_ylabel('test accuracy', fontsize=12);
ax[0][0].tick_params(labelsize=12);
ax[0][0].set_ylim([0.45, 0.85])
ax[0][0].set_yticks([0.5, 0.6, 0.7, 0.8]);
ax[0][0].set_xlabel('number of labeled example',fontsize=12)
ax[0][0].set_title('rep_size=50', fontsize=14)
ax[0][0].legend(loc='lower right', fontsize=12)

ax[0][1].plot(test_250['n_train'], test_250['word'])
ax[0][1].plot(test_250['n_train'], test_250['lsa'])
ax[0][1].plot(test_250['n_train'], test_250['combo'])
ax[0][1].set_ylabel('test accuracy', fontsize=12);
ax[0][1].tick_params(labelsize=12);
ax[0][1].set_ylim([0.45, 0.85])
ax[0][1].set_yticks([0.5, 0.6, 0.7, 0.8]);
ax[0][1].set_xlabel('number of labeled example',fontsize=12)
ax[0][1].set_title('rep_size=250', fontsize=14)
ax[0][1].legend(loc='lower right', fontsize=12)

ax[1][0].plot(test_500['n_train'], test_500['word'])
ax[1][0].plot(test_500['n_train'], test_500['lsa'])
ax[1][0].plot(test_500['n_train'], test_500['combo'])
ax[1][0].set_ylabel('test accuracy', fontsize=12);
ax[1][0].tick_params(labelsize=12);
ax[1][0].set_ylim([0.45, 0.85])
ax[1][0].set_yticks([0.5, 0.6, 0.7, 0.8]);
ax[1][0].set_xlabel('number of labeled example',fontsize=12)
ax[1][0].set_title('rep_size=500', fontsize=14)
ax[1][0].legend(loc='lower right', fontsize=12)

ax[1][1].plot(test_1000['n_train'], test_1000['word'])
ax[1][1].plot(test_1000['n_train'], test_1000['lsa'])
ax[1][1].plot(test_1000['n_train'], test_1000['combo'])
ax[1][1].set_ylabel('test accuracy', fontsize=12);
ax[1][1].tick_params(labelsize=12);
ax[1][1].set_ylim([0.45, 0.85])
ax[1][1].set_yticks([0.5, 0.6, 0.7, 0.8]);
ax[1][1].set_xlabel('number of labeled example',fontsize=12)
ax[1][1].set_title('rep_size=1000', fontsize=14)
ax[1][1].legend(loc='lower right', fontsize=12)
plt.tight_layout();
fig.savefig('./hw1_figures/part1_4.png', dpi=500);
```

```python
word_rep = pd.concat((test_50['n_train'], test_50['word'], test_250['word'], test_500['word'], 
                      test_750['word'], test_1000['word']), axis=1)
word_rep.columns = ['n_train', 'rep_size=50', 'rep_size=250', 'rep_size=500', 'rep_size=750', 'rep_size=1000']

lsa_rep = pd.concat((test_50['n_train'], test_50['lsa'], test_250['lsa'], test_500['lsa'], 
                      test_750['lsa'], test_1000['lsa']), axis=1)
lsa_rep.columns = ['n_train', 'rep_size=50', 'rep_size=250', 'rep_size=500', 'rep_size=750', 'rep_size=1000']

combo_rep = pd.concat((test_50['n_train'], test_50['combo'], test_250['combo'], test_500['combo'], 
                      test_750['combo'], test_1000['combo']), axis=1)
combo_rep.columns = ['n_train', 'rep_size=50', 'rep_size=250', 'rep_size=500', 'rep_size=750', 'rep_size=1000']
```

```python
word_rep.head()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(13, 4))
ax[0].plot(word_rep['n_train'], word_rep['rep_size=50'])
ax[0].plot(word_rep['n_train'], word_rep['rep_size=250'])
ax[0].plot(word_rep['n_train'], word_rep['rep_size=500'])
ax[0].plot(word_rep['n_train'], word_rep['rep_size=750'])
ax[0].plot(word_rep['n_train'], word_rep['rep_size=1000'])
ax[0].set_ylabel('test accuracy', fontsize=12);
ax[0].tick_params(labelsize=12);
ax[0].set_ylim([0.45, 0.85])
ax[0].set_yticks([0.5, 0.6, 0.7, 0.8]);
ax[0].set_xlabel('number of labeled example',fontsize=12)
ax[0].set_title('word', fontsize=14)
ax[0].legend(loc='lower right', fontsize=12)

ax[1].plot(lsa_rep['n_train'], lsa_rep['rep_size=50'])
ax[1].plot(lsa_rep['n_train'], lsa_rep['rep_size=250'])
ax[1].plot(lsa_rep['n_train'], lsa_rep['rep_size=500'])
ax[1].plot(lsa_rep['n_train'], lsa_rep['rep_size=750'])
ax[1].plot(lsa_rep['n_train'], lsa_rep['rep_size=1000'])
ax[1].set_ylabel('test accuracy', fontsize=12);
ax[1].tick_params(labelsize=12);
ax[1].set_ylim([0.45, 0.85])
ax[1].set_yticks([0.5, 0.6, 0.7, 0.8]);
ax[1].set_xlabel('number of labeled example',fontsize=12)
ax[1].set_title('lsa', fontsize=14)
ax[1].legend(loc='lower right', fontsize=12)

ax[2].plot(combo_rep['n_train'], combo_rep['rep_size=50'])
ax[2].plot(combo_rep['n_train'], combo_rep['rep_size=250'])
ax[2].plot(combo_rep['n_train'], combo_rep['rep_size=500'])
ax[2].plot(combo_rep['n_train'], combo_rep['rep_size=750'])
ax[2].plot(combo_rep['n_train'], combo_rep['rep_size=1000'])
ax[2].set_ylabel('test accuracy', fontsize=12);
ax[2].tick_params(labelsize=12);
ax[2].set_ylim([0.45, 0.85])
ax[2].set_yticks([0.5, 0.6, 0.7, 0.8]);
ax[2].set_xlabel('number of labeled example',fontsize=12)
ax[2].set_title('combo', fontsize=14)
ax[2].legend(loc='lower right', fontsize=12)

plt.tight_layout();
fig.savefig('./hw1_figures/part1_5.png', dpi=500);
```
